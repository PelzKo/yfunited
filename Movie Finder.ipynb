```python
import requests
import random
from IPython.display import YouTubeVideo
```

```python
api_key = "k_5sd7zrl1"
get_movies_url = "https://imdb-api.com/en/API/MostPopularMovies/"+api_key
get_trailer_url = "https://imdb-api.com/en/API/YouTubeTrailer/"+api_key
```

```python
def get_request(url,id=""):
    payload = {}
    headers= {}
    response = requests.request("GET", url+"/"+id, headers=headers, data = payload)
    return response.json()
```

```python
#response = requests.request("GET", get_movies_url, headers=headers, data = payload)
#print(response.text.encode('utf8'))

movies = get_request(get_movies_url).get("items")
```

```python
random_index = random.randint(0,len(movies)-1)

chosen_movie = movies[random_index]
print(chosen_movie["title"])
```

```python
trailer_id = get_request(get_trailer_url,chosen_movie["id"]).get("videoId")
print(trailer_id)
```
